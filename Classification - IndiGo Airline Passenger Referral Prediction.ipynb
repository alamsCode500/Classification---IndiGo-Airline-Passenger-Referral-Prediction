## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

try:
    # Data Wrangling Process

    # Step 1: Drop duplicate rows
    dataset = dataset.drop_duplicates()
    print("Duplicate rows removed.")

    # Step 2: Handle missing values (example: drop rows with missing 'airline' or 'author')
    dataset = dataset.dropna(subset=['airline', 'author']).reset_index(drop=True)
    print("Rows with missing 'airline' or 'author' dropped.")

    # Step 3: Convert 'review_date' and 'date_flown' to datetime format
    dataset['review_date'] = pd.to_datetime(dataset['review_date'], errors='coerce')
    dataset['date_flown'] = pd.to_datetime(dataset['date_flown'], errors='coerce')
    print("Date columns converted to datetime format.")

    # Step 4: Fill missing numeric columns with their median values
    numeric_cols = dataset.select_dtypes(include=['float64', 'int64']).columns
    dataset[numeric_cols] = dataset[numeric_cols].fillna(dataset[numeric_cols].median())
    print("Missing values in numeric columns filled with median values.")

    # Step 5: Fill missing categorical columns with mode values
    categorical_cols = dataset.select_dtypes(include=['object']).columns
    dataset.update(dataset[categorical_cols].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown')))
    print("Missing values in categorical columns filled with mode values.")

    # Step 6: Cleanup for text data (remove special characters in 'customer_review')
    dataset['customer_review'] = dataset['customer_review'].str.replace(r'[^\w\s]', '', regex=True)
    print("Special characters removed from 'customer_review'.")

    # Step 7: Display first 5 rows of the dataset
    print("\nFirst 5 rows of the dataset:")
    print(dataset.head())

except Exception as e:
    # Handle any errors that occur
    print(f"An error occurred during the data wrangling process: {e}")

finally:
    # Confirm that the data wrangling process is completed
    print("Data wrangling process completed.")
````

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

**Hypothetical Statements:**

**Hypothesis 1:** Passengers who rated Cabin Service higher are more likely to recommend the airline.

**Hypothesis 2:** There is a significant difference in the Overall Rating given by Solo Travelers and Family Leisure Travelers.

**Hypothesis 3:** Airlines with a higher Seat Comfort rating also have a higher Overall Rating.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H0):**

Passengers who rated Cabin Service higher are not more likely to recommend the airline. There is no significant difference in the mean Cabin Service ratings between passengers who recommended the airline and those who did not.

**Alternate Hypothesis (Ha):**

Passengers who rated Cabin Service higher are more likely to recommend the airline. There is a significant difference in the mean Cabin Service ratings between passengers who recommended the airline and those who did not.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

from scipy.stats import ttest_ind

try:
    # Split the dataset into two groups: Recommended and Not Recommended
    recommended_group = dataset[dataset['recommended'] == 'yes']['cabin_service']
    not_recommended_group = dataset[dataset['recommended'] == 'no']['cabin_service']

    # Perform an independent t-test
    t_stat, p_value = ttest_ind(recommended_group, not_recommended_group, nan_policy='omit')

    # Print the results
    print(f"T-Statistic: {t_stat}")
    print(f"P-Value: {p_value}")

    # Determine statistical significance
    alpha = 0.05
    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis. Passengers who rated Cabin Service higher are more likely to recommend the airline.")
    else:
        print("Conclusion: Fail to reject the null hypothesis. No evidence suggests that Cabin Service rating impacts recommendation likelihood.")

except Exception as e:
    print(f"An error occurred during the statistical test: {e}")

finally:
    print("Statistical test for Hypothesis 1 completed.")
````

##### Which statistical test have you done to obtain P-Value?

The statistical test performed to obtain the P-value is an **independent t-test** (also called a two-sample t-test).

##### Why did you choose the specific statistical test?

**Why the Independent t-Test?**

**1. Purpose:**

The independent t-test is used to compare the means of a numeric variable between two independent groups. In this case, it checks if passengers who recommend the airline have significantly different mean Cabin Service ratings compared to those who do not.

**2. Type of Data:**

* **Cabin Service Rating:** This is a continuous numeric variable.

* **Recommended:** This is a binary categorical variable with two independent groups:

* **Group 1:** Passengers who recommended the airline (e.g., yes).

* **Group 2:** Passengers who did not recommend the airline (e.g., no).

**3. Assumptions:**

The t-test was chosen because:

* The data is approximately normally distributed for each group (assumption of normality).

* The two groups (recommended vs. not recommended) are independent.

* The variances of the Cabin Service ratings are either assumed to be equal (pooled t-test) or checked with a variance test (Welch’s t-test if variances are unequal).

**4. Hypothesis:**

* **Null Hypothesis (H₀):** The mean Cabin Service ratings are the same for both groups.

* **Alternative Hypothesis (Hₐ):** The mean Cabin Service ratings are different for the two groups.

**5. Result:**

* **t-statistic:** Indicates the magnitude and direction of the difference in means.

* **P-value:** Helps to decide if the observed difference in means is statistically significant.

By using the t-test, we can objectively determine if Cabin Service ratings influence a passenger's likelihood to recommend the airline.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H₀):**

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments

try:

    # Boxplot to visualize outliers Before treatment
    plt.figure(figsize=(12, 6))
    for i, col in enumerate(numeric_cols[:7]):  # Plotting numeric columns for visualization
        plt.subplot(4, 2, i + 1)
        sns.boxplot(x=dataset[col])
        plt.title(f"Boxplot of {col} Before Treatment")
    plt.tight_layout()
    plt.show()

    # Identifying numeric columns
    numeric_cols = dataset.select_dtypes(include=['float64', 'int64']).columns

    # Identifying outliers using the IQR method
    Q1 = dataset[numeric_cols].quantile(0.25)  # First quartile (25th percentile)
    Q3 = dataset[numeric_cols].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile Range

    # Defining outlier boundaries
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Finding outliers
    outliers = ((dataset[numeric_cols] < lower_bound) | (dataset[numeric_cols] > upper_bound)).sum()
    print(f"Outliers Count Before Treatment:\n{outliers}")

    # Outlier Treatment
    for col in numeric_cols:
        dataset[col] = np.where(dataset[col] < lower_bound[col], lower_bound[col], dataset[col])
        dataset[col] = np.where(dataset[col] > upper_bound[col], upper_bound[col], dataset[col])

    # Outliers after treatment
    new_outliers = ((dataset[numeric_cols] < lower_bound) | (dataset[numeric_cols] > upper_bound)).sum()
    print(f"\nOutliers Count After Treatment:\n{new_outliers}")

    # Boxplot to visualize outliers after treatment
    plt.figure(figsize=(12, 6))
    for i, col in enumerate(numeric_cols[:7]):  # Plotting numeric columns for visualization
        plt.subplot(4, 2, i + 1)
        sns.boxplot(x=dataset[col])
        plt.title(f"Boxplot of {col} After Treatment")
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"An error occurred during outlier handling: {e}")

finally:
    print("Outlier treatment process completed.")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

**IQR Method (Interquartile Range)**

* This method detects outliers by identifying values that fall below Q1 - 1.5 × IQR or above Q3 + 1.5 × IQR.

* It is useful for handling non-normally distributed data and reduces the effect of extreme values without being affected by mean or standard deviation.

**Winsorization (Capping Method)**

* Instead of removing outliers, extreme values are replaced with the upper and lower IQR boundaries.

* This ensures that data integrity is preserved while reducing the impact of extreme values.

**Boxplot Visualization**

* Before and after applying outlier treatments, boxplots are used to visually confirm the presence and removal of outliers.

* This helps in understanding how much the data was affected by extreme values and whether the treatment was effective.

**Why These Techniques?**
* IQR is robust because it is based on quartiles rather than mean and standard deviation, making it ideal for skewed data.

* Winsorization prevents data loss and keeps extreme values in check without completely removing them.

* Boxplots provide a visual check to ensure that the outlier treatment method was applied correctly.

````python
# Rephrase Text

# A dictionary of common misspellings or common words to rephrase
misspellings = {
    "alot": "a lot",
    "definately": "definitely",
    "seperately": "separately",
    "occured": "occurred",
    "recieve": "receive",
    "untill": "until",
    "whould": "would"
}

# Function to rephrase text (lowercase, handle basic corrections, and whitespace)
def rephrase_text(text):
    # Lowercase the text
    text = text.lower()

    # Replace common misspellings
    for word, correct_word in misspellings.items():
        text = text.replace(word, correct_word)

    # Remove multiple spaces
    text = re.sub(r'\s+', ' ', text)

    # Trim leading and trailing spaces
    text = text.strip()

    return text

try:
    # Ensure 'customer_review' column exists in the dataset
    if 'customer_review' in dataset.columns:
        # Apply the rephrasing function to each row in 'customer_review'
        dataset['customer_review'] = dataset['customer_review'].apply(rephrase_text)
        print("Text in 'customer_review' rephrased successfully.")
    else:
        print("'customer_review' column not found in the dataset, skipping text rephrasing.")

except Exception as e:
    print(f"An error occurred during text rephrasing: {e}")

finally:
    print("Text rephrasing process completed.")
````

````python
# Step 3: Create New Features

from textblob import TextBlob

try:
    # Generate text-based features
    dataset_copy["review_length"] = dataset_copy["customer_review"].astype(str).apply(len)
    dataset_copy["word_count"] = dataset_copy["customer_review"].astype(str).apply(lambda x: len(x.split()))

    # Sentiment Score
    dataset_copy["sentiment"] = dataset_copy["customer_review"].astype(str).apply(lambda x: TextBlob(x).sentiment.polarity)

    if all(col in dataset_copy.columns for col in ["seat_comfort", "cabin_service", "ground_service"]):
        dataset_copy["service_avg"] = (
            dataset_copy["seat_comfort"] + dataset_copy["cabin_service"] + dataset_copy["ground_service"]
        ) / 3

    print("New features added successfully.")

except Exception as e:
    print(f"An error occurred during feature engineering: {e}")

finally:
    print("Feature engineering completed.")
````

````python
# Select your features wisely to avoid overfitting

from sklearn.feature_selection import mutual_info_classif
from sklearn.preprocessing import LabelEncoder

try:
    print("Initial columns:", dataset_copy.columns.tolist())

    # Feature processing on text columns such as `customer_review_tokens`, `customer_review_tokens_stemmed`, etc.
    text_cols = ['customer_review', 'customer_review_tokens', 'customer_review_tokens_stemmed', 'customer_review_text']

    # Process Numerical Features ----
    numeric_cols = dataset_copy.select_dtypes(include=['number']).drop(columns=['recommended'], errors='ignore')

    # Process Categorical Features ----
    categorical_cols = dataset_copy.select_dtypes(include=['object']).drop(columns=text_cols, errors='ignore')

    # Encoding Categorical Features
    categorical_encoded = pd.DataFrame()
    for col in categorical_cols.columns:
        encoder = LabelEncoder()
        dataset_copy[col] = encoder.fit_transform(dataset_copy[col].astype(str))
        categorical_encoded[col] = dataset_copy[col]

    # Process PCA and SVD Components ----
    pca_svd_features = [col for col in dataset_copy.columns if 'PCA_' in col or 'SVD_' in col]

    # Combine all numerical, encoded categorical, and text-based features
    X = pd.concat([numeric_cols, categorical_encoded], axis=1)

    # Apply Mutual Information for Feature Selection ----
    y = dataset_copy["recommended"]

    if y is not None and not X.empty:
        mi_scores = mutual_info_classif(X, y, discrete_features="auto")
        mi_scores = dict(zip(X.columns, mi_scores))

        # Select Top Features (e.g., Top 20)
        selected_features = sorted(mi_scores, key=mi_scores.get, reverse=True)[:20]
        dataset_copy = dataset_copy[selected_features + ['recommended']]

        print(f"Top 20 important features selected: {selected_features}")

except Exception as e:
    print(f"An error occurred during feature selection: {e}")

finally:
    print("Feature selection completed.")
````

* **No Assumption of Linear Relationships:** Unlike some methods (e.g., correlation-based), MI does not assume that the relationships between features and the target are linear, making it more flexible.

* **Incorporates Non-linear Relationships:** MI captures more complex relationships that might be missed by linear methods.

**Model Performance:**

**Reasons Dimensionality Reduction Might Not Be Needed:**

**Sufficient Data:**

If the dataset has a relatively small number of features and a large number of observations, dimensionality reduction might not be necessary. The model can handle high-dimensional data well without overfitting.

**Important Features:**

If the features are independent and contain important information, removing them through dimensionality reduction could result in loss of valuable data.

#### 3. Explain each evaluation metric's indication towards business and the business impact of the ML model used.

**1. Accuracy**

* **Indication:** Accuracy indicates the proportion of correct predictions (both true positives and true negatives) out of all predictions made by the model. It provides a simple overall measure of the model's performance.

* **Business Impact:**

* In a business context, high accuracy means the model is correctly predicting outcomes (e.g., whether a customer will refer the airline or not) most of the time.

* However, accuracy alone might not be sufficient if the dataset is imbalanced (e.g., if one class dominates the dataset), as the model might predict the majority class well but fail on the minority class.

* **Business Relevance:** For customer experience prediction, achieving high accuracy would help in identifying the customers who might not refer the airline, allowing the business to intervene and improve customer satisfaction.

**2. Precision**

* **Indication:** Precision is the ratio of true positives (correctly predicted positive class) to all predicted positives (sum of true positives and false positives). It reflects the ability of the model to avoid false positives.

* **Business Impact:**

* High precision means fewer false positives, which is critical when the cost of a false positive is high. For example, in a referral prediction model, a false positive would mean incorrectly predicting that a customer will refer the airline when they will not.

* **Business Relevance:** Minimizing false positives ensures that marketing efforts are not wasted by targeting customers who are unlikely to refer the airline. This helps improve cost-efficiency in marketing campaigns and enhances customer targeting.

**3. Recall**

* **Indication:** Recall, also known as sensitivity or true positive rate, is the ratio of true positives to all actual positives (sum of true positives and false negatives). It reflects the model's ability to correctly identify all relevant instances.

* **Business Impact:**

* High recall means fewer false negatives, which is important when the cost of missing a positive instance is high. In the case of predicting customer referrals, a false negative would be missing out on a customer who was actually going to refer the airline.

* **Business Relevance:** For a business, maximizing recall helps in identifying all customers who are likely to refer the airline, ensuring that no potential promoters are overlooked, which can increase the likelihood of positive word-of-mouth marketing.

**4. F1-Score**

* **Indication:** The F1-score is the harmonic mean of precision and recall, providing a single score that balances both the concerns of precision and recall. It is especially useful when dealing with imbalanced datasets.

* **Business Impact:**

* A high F1-score indicates that the model performs well in both identifying relevant instances (recall) and minimizing false positives (precision).

* **Business Relevance:** In scenarios like referral prediction, a good F1-score ensures that the business can target the right customers while maintaining cost efficiency. It ensures that the efforts made to increase customer referrals are optimized for both accuracy and efficiency.

**5. Support**

* **Indication:** Support refers to the number of actual occurrences of each class in the dataset. It gives insight into the distribution of the classes in the data.

* **Business Impact:**

* Understanding the support helps the business recognize which class (e.g., customers who will refer vs. those who won't) is more prevalent, guiding strategy decisions.

### 3. Explain the model which you have used and the feature importance using any model explainability tool?

**Model Used: Random Forest Classifier**

The Random Forest Classifier is an ensemble learning method that operates by creating multiple decision trees during training and merging their outputs to improve the accuracy and robustness of predictions. It is known for its ability to handle both regression and classification tasks efficiently and effectively. Random Forest mitigates overfitting by combining results from many trees, each trained on different subsets of the data. The final prediction is made based on the majority vote from all the trees in the forest, ensuring more stable and generalizable results.

**Feature Importance in Random Forest**

Feature importance refers to a technique that helps identify which features are most impactful in the model's decision-making process. In Random Forest, each decision tree is split based on the features that reduce impurity, such as the Gini impurity or entropy in classification tasks. The more a feature contributes to reducing the impurity in the trees, the more important it is considered.

* **How Feature Importance is Calculated:**

* **Impurity Reduction:** Random Forest evaluates each feature based on how much it reduces the impurity (Gini or entropy) at every split in each tree. Features that lead to more significant impurity reductions are given higher importance.

* **Averaging Across Trees:** Since Random Forest is an ensemble of multiple trees, the importance of each feature is averaged across all the trees in the forest.

* **Normalization:** The importance of each feature is normalized, meaning the sum of all feature importances will be 1. This helps in understanding the relative contribution of each feature.

**Model Explainability Tools for Feature Importance**

To better understand the importance of features in Random Forest models, several model explainability tools can be used:

* **SHAP (SHapley Additive exPlanations):** SHAP is a popular technique that uses game theory to determine the contribution of each feature towards a specific prediction. SHAP values help in interpreting how each feature influences the outcome, making them highly useful for explaining individual predictions and understanding global model behavior.

* **Permutation Importance:** This method calculates feature importance by randomly shuffling the values of a feature and measuring the decrease in the model's performance. Features that, when shuffled, cause the largest drop in performance are considered more important.

* **Sklearn’s feature_importances_:** Random Forest in Scikit-learn has a built-in method to access feature importance directly from the trained model. It ranks features based on how much they contribute to reducing impurity across the decision trees in the forest.
